````python
# %load drawcircle.py
import math
import turtle

# draw the circle using turtle
def drawCircleTurtle(x, y, r):
	# move to the start of the circle
	turtle.up()
	turtle.setpos(x + r, y)
	turtle.down()

	# draw the circle
	for i in range(0, 365, 5):
		a = math.radians(i)
		turtle.setpos(x + r * math.cos(a), y + r * math.sin(a))

drawCircleTurtle(100, 100, 50)
turtle.mainloop()
````
